```python
def IQR(x):
  q1=x.quantile(0.25)
  q3=x.quantile(0.75)
  iqr=q3-q1
  return iqr
```

```python
df.drop(columns={"PoolArea","PoolQC","Fence","MiscFeature"},inplace=True)
```

```python
def Upper(x):
  q3=x.quantile(0.75)
  UF=q3+1.5*IQR(x)
  return UF
```

```python
def Lower(x):
  q1=x.quantile(0.25)
  LF=q1-1.5*IQR(x)
  return LF
```

```python
for column in df.columns:
    if df[column].dtype == 'object':
        values = df[column].unique()
        mapping = {}
        i = 0
        for val in values:
            if pd.notna(val):
                mapping[val] = i
                i += 1
        df[column] = df[column].map(mapping)
```
